```python
import pandas as pd
import statsmodels.api as sm
# Read in the data and fit the model
d = pd.read_csv("derbydata.csv")

model = sm.OLS.from_formula("speed ~ year", data = d)
result = model.fit()
result.summary()
```

```python
# import and run diagnostic plots
from residualplots import diagnostic_plots

diagnostic_plots(result, y = d.speed)
```
